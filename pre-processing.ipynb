# Cleaning the Data

```python
clothes_df_clean = clothes_df.copy()
```

## Cleaning the Numerical Features

```python
clothes_df_clean.rename({'DiscountPrice (in Rs)': 'DiscountPrice',
                         "OriginalPrice (in Rs)": "OriginalPrice"},
                        axis=1,
                        inplace=True)
```

```python
clothes_df_clean.head()
```

```python
clothes_df_clean.columns
```

```python
# Data type correction
clothes_df_clean['DiscountPrice'] = clothes_df_clean['DiscountPrice'].astype(float)
clothes_df_clean['OriginalPrice'] = clothes_df_clean['OriginalPrice'].astype(float)
clothes_df_clean['Ratings'] = clothes_df_clean['Ratings'].astype(float)
clothes_df_clean['Reviews'] = clothes_df_clean['Reviews'].astype(int)


# Extract numeric values from DiscountOffer - just incase the split didn't
clothes_df_clean['DiscountOffer'] = clothes_df_clean['DiscountOffer'].str.extract('(\d+)').astype(float)
```

## Checking the distribtution of the numerical featuers

```python
numerical_features = ['DiscountPrice', 'OriginalPrice', 'Ratings', 'Reviews', 'DiscountOffer']
for feature in numerical_features:
  plt.figure(figsize=(10, 6))
  sns.histplot(clothes_df_clean[feature], bins=30, kde=True)
  plt.title(f'Distribution of {feature}')
  plt.show()
```

## Cleaning the Categorical Features

```python
# Extract information from URL
clothes_df_clean['product_name'] = clothes_df_clean['URL'].apply(lambda x: x.split('/')[-3])
```

```python
clothes_df_clean['product_name'].sample(5)
```

```python
from sklearn.preprocessing import OneHotEncoder, OrdinalEncoder
from sklearn.decomposition import PCA
```

### BrandName and Individual_category

```python
# One-Hot Encoding for BrandName and Individual_category
# Initialize separate encoders
brand_encoder = OneHotEncoder(sparse_output=False, drop='first')
category_encoder = OneHotEncoder(sparse_output=False, drop='first')
```

```python
# Fit and transform
encoded_brandname = brand_encoder.fit_transform(clothes_df_clean[['BrandName']])
encoded_individual_category = category_encoder.fit_transform(clothes_df_clean[['Individual_category']])
```

```python
len(encoded_brandname)
```

```python
len(encoded_individual_category)
```

```python
# Convert to DataFrames using the correct encoders --brandname
encoded_brandname_df = pd.DataFrame(encoded_brandname, columns=brand_encoder.get_feature_names_out(['BrandName']))
```

```python
encoded_brandname_df.shape
```

```python
# Convert to DataFrames using the correct encoders --individual category
encoded_individual_category_df = pd.DataFrame(encoded_individual_category, columns=category_encoder.get_feature_names_out(['Individual_category']))
```

```python
encoded_individual_category_df.shape
```

```python
encoded_individual_category_df.head()
```

```python
pca_brandname = PCA(n_components=20)
pca_individual_category = PCA(n_components=20)
```

```python
pca_brandname_features = pca_brandname.fit_transform(encoded_brandname_df)
```

```python
len(pca_brandname_features)
```

```python
pca_individual_category_features = pca_individual_category.fit_transform(encoded_individual_category_df)
```

```python
len(pca_individual_category_features)
```

```python
pca_brandname_df = pd.DataFrame(pca_brandname_features, columns=[f'pca_brandname_{i}' for i in range(20)])
```

```python
pca_brandname_df
```

```python
pca_individual_category_df = pd.DataFrame(pca_individual_category_features, columns=[f'pca_individual_category_{i}' for i in range(20)])
```

```python
pca_individual_category_df
```

```python
# Drop original columns and concatenate PCA features
clothes_df_clean = clothes_df_clean.drop(columns=['BrandName', 'Individual_category'])
clothes_df_clean = pd.concat([clothes_df_clean, pca_brandname_df, pca_individual_category_df], axis=1)
```

```python
clothes_df_clean.shape
```

```python
## Category and Category_by_Gender
```

```python
# Separate encoders for each feature
encoder_category = OneHotEncoder(sparse_output=False, drop='first')
encoder_gender = OneHotEncoder(sparse_output=False, drop='first')
```

```python
# Fit and transform the 'Category' feature
encoded_category = encoder_category.fit_transform(clothes_df_clean[['Category']])
```

```python
encoded_category_df = pd.DataFrame(encoded_category, columns=encoder_category.get_feature_names_out(['Category']))
```

```python
encoded_category_df
```

```python
# Fit and transform the 'category_by_Gender' feature
encoded_gender = encoder_gender.fit_transform(clothes_df_clean[['category_by_Gender']])
```

```python
encoded_gender_df = pd.DataFrame(encoded_gender, columns=encoder_gender.get_feature_names_out(['category_by_Gender']))
```

```python
encoded_category_df.shape
```

```python
# Drop original columns and concatenate encoded features
# Check if columns exist before dropping
if 'Category' in clothes_df_clean.columns and 'category_by_Gender' in clothes_df_clean.columns:
    clothes_df_clean = clothes_df_clean.drop(columns=['Category', 'category_by_Gender'])

clothes_df_clean = pd.concat([clothes_df_clean, encoded_category_df, encoded_gender_df], axis=1)
```
